```python
%load_ext autoreload
%autoreload 2
%load_ext line_profiler

import numpy as np
import matplotlib.pyplot as plt

from ml.apps.skin.timeseries import *
from ml.apps.skin.workingset import *
from ml.simple.table import *
from ml.datasets.skin import SkinDataset
from ml.common.util import ParameterHistory
from ml.common.test import check_gradient, check_directional_gradient
from climin.gd import GradientDescent
from climin.bfgs import Bfgs, Lbfgs
from climin.rmsprop import RmsProp


ws = SkinWorkingset("raising_small", (1,1))
```

# Pretraining using matrix inversion

```python
tr = SmoothTableRegression([ws.force_min, ws.skin_min], 
                           [ws.force_step, ws.skin_step], 
                           [ws.force_max, ws.skin_max])
print "Number of weights: %d" % tr.weight_matrix.size

start_weights = np.load("skin-table-start-weights.npz")['optimal_weights']
tr.weights = start_weights.copy()

err = ws.error(tr)
ws.print_error(err)
```

## Finetuning using iterative full prediction and gradient descent

```python
new_batch_iters = 10
restart_steps = 10
       
his_failed = ParameterHistory(min_iters=1000)
his_ns = ParameterHistory(show_progress=False)
his_ms = ParameterHistory(show_progress=False)            

msws = SkinMultistepWorkingset(ws, tr, ws.force['trn'], ws.skin['trn'], ws.valid['trn'], 50)
nsws = SkinNextstepWorkingset(ws, tr, None, None)
opt = RmsProp(tr.weights, nsws.gradient_wrapper, steprate=1e-5)

msws.batch = 0
f, v, s = msws.get_fvs()
s_p = multistep_predict(tr.predict, f, v, s[0, :])
nsws.ns_in, nsws.ns_skin = build_nextstep_data_from_multicurve(f, s_p, v, s)

for iter, info in enumerate(opt):
    if iter % new_batch_iters == 0:
        msws.next_batch()
        f, v, s = msws.get_fvs()
        if restart_steps == 0:
            s_p = multistep_predict(tr.predict, f, v, s[0, :])
        else:
            s_p = restarting_multistep_predict(tr.predict, f, v, s, restart_steps)
        nsws.ns_in, nsws.ns_skin = build_nextstep_data_from_multicurve(f, s_p, v, s)

    if iter % 100 == 0:        
        err = ws.error(tr)
        his_ns.add(iter, tr.weights, err['ns_trn'], err['ns_val'], err['ns_tst'])
        his_ms.add(iter, tr.weights, err['ms_trn'], err['ms_val'], err['ms_tst'])
        his_failed.add(iter, tr.weights, len(err['failed_trn']), len(err['failed_val']), len(err['failed_tst']))                  
        
        plt.clf()
        his_failed.plot(final=False, logscale=False)
        plt.show()           
    
        if len(err['failed_trn']) < 1:
            break
            
np.savez_compressed("multistep-iterative.npz", weights=tr.weights)         
```
