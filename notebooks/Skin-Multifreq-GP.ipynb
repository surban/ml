## Training

```python
kernel = gpy.kern.rbf(ws.n_features)
X = ws.flat_skin['trn'].T
Y = ws.flat_force['trn'][:, np.newaxis]
gp = gpy.models.GPRegression(X, Y, kernel)

gp.optimize()
print gp
```

```python
p_flat_force, p_flat_force_var, _, _ = gp.predict(ws.flat_skin['tst'].T, full_cov=False)

figsize(20,10)
plt.plot(ws.flat_force['tst'], label='truth')
plt.plot(p_flat_force, label='prediction')
plt.legend()
```

```python
prt = 'val'
n_plot_samples = 30

force = ws.force[prt][:, 0:n_plot_samples]
skin = ws.skin[prt][:, :, 0:n_plot_samples]
valid = ws.valid[prt][:, 0:n_plot_samples]

def gp_predict(skin):
    force, force_conf, _, _ = gp.predict(skin.T, full_cov=False)
    return force[:, 0], force_conf[:, 0]

p_force, p_force_conf = ws.predict_multicurve(gp_predict, force, skin, valid, what='force')

ml.apps.skin.timeseries.max_force = 10
plot_multicurve_time(force, None, valid, 
                     force_predicted=p_force, force_predicted_conf=None)
```
