```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt

from ml.simple.vargp import VarGP
```

```python
def myfunc(x):
    mean = np.sin(x)
    std = 0.1
    val = mean + np.random.normal(size=x.shape) * std
    return val
```

```python
removep = True
```

```python
n_samples = 50
max_x = 4.5*np.pi
x = np.random.uniform(low=0, high=max_x, size=n_samples)
```

```python
if removep:
    x = x[np.logical_or(x<6, x>9)]
y = myfunc(x)

figsize(8,4)
plt.plot(x,y, '.')
plt.xlim(0, max_x)
plt.ylim(-1.5, 1.5)
```

```python
vgp = VarGP(x[np.newaxis, :], y, optimize_restarts=5, use_data_variance=False)
print vgp
```

```python
figsize(20,10)
tst_x = np.linspace(0, max_x, 100)
vgp.plot(x[np.newaxis, :], y, tst_x[np.newaxis, :])
```
