```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
import GPy as gpy

import ml.apps.skin.timeseries
from ml.apps.skin.timeseries import *
from ml.apps.skin.workingset import *
from ml.simple.conditional_chain import *
from ml.datasets.skin import SkinDataset

ml.apps.skin.timeseries.max_force = 10
```

## Dataset and features selection

```python
def select_features(ws, skin):    
    return ws.select_frequencies([794328, 2511886], skin)

force_min = -0.2
force_max = 10.0

ws_linear_bidir = SkinWorkingset("mf_linear_bidir_10", taxel=(0,6), force_min=force_min, force_max=force_max)
ws_linear_raising = SkinWorkingset("mf_linear_raising_10", taxel=(0,6), force_min=force_min, force_max=force_max)

ws_random_raising = SkinWorkingset("mf_random_raising_10", taxel=(0,6), force_min=force_min, force_max=force_max)
ws_random_bidir = SkinWorkingset("mf_random_bidir_10", taxel=(0,6), force_min=force_min, force_max=force_max)

ws_hysteresis_10 = SkinWorkingset("mf_hysteresis_10", taxel=(0,6), force_min=force_min, force_max=force_max)
ws_hysteresis_50 = SkinWorkingset("mf_hysteresis_50", taxel=(0,6), force_min=force_min, force_max=force_max)
ws_hysteresis_100 = SkinWorkingset("mf_hysteresis_100", taxel=(0,6), force_min=force_min, force_max=force_max)
```

## Pretraining of observation -> state model using hysteresis-free data

```python
ws_raising_trn = ws_linear_raising

states = ws_raising_trn.discrete_force['trn']
observations = select_features(ws_raising_trn, ws_raising_trn.skin['trn'])
valid = ws_raising_trn.valid['trn']
n_features = observations.shape[0]

sc = ControlObservationChain(ws_raising_trn.discrete_force_states, ws_raising_trn.discrete_force_states)
sc.train_observations(states, observations, valid,                      
                      kernel=lambda: gpy.kern.rbf(n_features),
                      optimize_restarts=10)  
print sc.gp_states_given_observations
```

#### Evaluation

```python
def evaluate_on_ws(sc, ws, prt, phase=False):
    n_plot_samples = 30

    force = ws.force[prt]
    skin = select_features(ws, ws.skin[prt])
    valid = ws.valid[prt]

    p_force_discr, p_force_conf_discr = ws.predict_multicurve(sc.infer_states, force, skin, valid, what='force')
    p_force = ws.from_discrete_force(p_force_discr)
    p_force_conf = p_force_conf_discr * ws.force_step
    print "MSE: ", multistep_error(p_force, force, valid, mean_err=True)
    
    ml.apps.skin.timeseries.max_force = 10
    if phase:
        plot_multicurve_error(force[:, 0:n_plot_samples], p_force[:, 0:n_plot_samples], valid[:, 0:n_plot_samples])
    else:
        plot_multicurve_time(force[:, 0:n_plot_samples], None, valid[:, 0:n_plot_samples], 
                             force_predicted=p_force[:, 0:n_plot_samples], force_predicted_conf=p_force_conf[:, 0:n_plot_samples])    
        
figsize(16, 8)
evaluate_on_ws(sc, ws_hysteresis_10, 'trn')
```

## Training of state -> state model using data with hysteresis

```python
ws_hys_trn = ws_hysteresis_100

discrete_force = ws_hys_trn.discrete_force['trn']
observations = select_features(ws_hys_trn, ws_hys_trn.skin['trn'])
valid = ws_hys_trn.valid['trn']

states, _ = ws_hys_trn.predict_multicurve(sc.infer_states, discrete_force, observations, valid, what='force')
states = np.asarray(states, dtype='int')

sc.train_transitions(states, discrete_force, valid, 
                     mngful_dist=4, 
                     kernel=lambda: gpy.kern.rbf(1) + gpy.kern.linear(1), 
                     optimize_restarts=3)                     

sc.check_normalization()
```

#### Evaluation: states from inputs

```python
ws_hys_tst = ws_hysteresis_100
prt = 'tst'

discrete_force = ws_hys_tst.discrete_force[prt]
observations = select_features(ws_hys_tst, ws_hys_tst.skin[prt])
valid = ws_hys_tst.valid[prt]

states, _ = ws_hys_tst.predict_multicurve(sc.infer_states, discrete_force, observations, valid, what='force')
p_states = ws_hys_tst.predict_multicurve(lambda inp: sc.most_probable_states_given_inputs(inp)[0], 
                                         discrete_force, states, valid, what='skin')

print "MSE: ", multistep_error(ws_hys_tst.from_discrete_force(p_states), 
                               ws_hys_trn.from_discrete_force(states), valid, mean_err=True)

n_plot_smpls = 42
figsize(18,14)
ml.apps.skin.timeseries.max_force = 10
force_real = ws_hys_tst.from_discrete_force(discrete_force[:, 0:n_plot_smpls])
plot_multicurve_time(ws_hys_tst.from_discrete_force(states[:, 0:n_plot_smpls]), None, valid[:, 0:n_plot_smpls], 
                     force_predicted=ws_hys_tst.from_discrete_force(p_states[:, 0:n_plot_smpls]), 
                     force_real=force_real,
                     timestep=ws_hys_tst.ds.timestep)
```

#### Evalulation: states and inputs from observations

```python
force = ws_hys_tst.force[prt]
observations = select_features(ws_hys_tst, ws_hys_tst.skin[prt])
valid = ws_hys_tst.valid[prt]

p_states, p_discrete_force = ws_hys_tst.predict_multicurve(lambda inp: sc.most_probable_states_and_inputs_given_observations(inp)[0:2], 
                                                           force, observations, valid, what='force')
p_force = ws_hys_tst.from_discrete_force(p_discrete_force)
p_internal_force = ws_hys_tst.from_discrete_force(p_states)

print "MSE: ", multistep_error(p_force, force, valid, mean_err=True)

n_plot_smpls = 42
figsize(18,14)
ml.apps.skin.timeseries.max_force = 10
plot_multicurve_time(force[:, 0:n_plot_smpls], None, valid[:, 0:n_plot_smpls], 
                     force_predicted=p_force[:, 0:n_plot_smpls], 
                     force_real=p_internal_force[:, 0:n_plot_smpls],
                     timestep=ws_hys_tst.ds.timestep)
```

## Infer true force from internal force (with smoothing prior with limited future horizon)

```python
sigma = 10
future_steps = 1

def predict_from_past(internal_force):
    n_steps = internal_force.shape[0]
    pred = np.zeros(n_steps)
    for step in range(n_steps):
        internal_force_window = internal_force[0:step+future_steps+1]
        p_force = sc.most_probable_inputs(internal_force_window, sigma=sigma)[0]
        pred[step] = p_force[step]
    return pred
    
force, skin, valid = ws_hys_tst.force[prt], ws_hys_tst.skin[prt], ws_hys_tst.valid[prt]
internal_force, _ = ws_hys_tst.predict_multicurve(gp_predict, force, select_features(ws_hys_tst, skin), valid, what='force')

discrete_force = ws_hys_trn.to_discrete_force(force)
discrete_internal_force = ws_hys_trn.to_discrete_force(internal_force)

discrete_force_p_smooth = ws_hys_tst.predict_multicurve(predict_from_past, 
                                                        discrete_force, discrete_internal_force, valid, what='force')
print "MSE inferred smooth: ", multistep_error(ws_hys_trn.from_discrete_force(discrete_force_p_smooth), 
                                               ws_hys_trn.from_discrete_force(discrete_force), valid, mean_err=True)
print "MSE inferred raw:    ", multistep_error(ws_hys_trn.from_discrete_force(discrete_force_p), 
                                               ws_hys_trn.from_discrete_force(discrete_force), valid, mean_err=True)
print "MSE internal:        ", multistep_error(ws_hys_trn.from_discrete_force(discrete_internal_force), 
                                               ws_hys_trn.from_discrete_force(discrete_force), valid, mean_err=True)

n_plot_smpls = 42
figsize(18,14)
ml.apps.skin.timeseries.max_force = 10
force_blue = ws_hys_trn.from_discrete_force(discrete_force_p[:, 0:n_plot_smpls])
force_blue_int = ws_hys_trn.from_discrete_force(discrete_internal_force[:, 0:n_plot_smpls])
plot_multicurve_time(ws_hys_trn.from_discrete_force(discrete_force[:, 0:n_plot_smpls]), None, valid[:, 0:n_plot_smpls], 
                     force_predicted=ws_hys_trn.from_discrete_force(discrete_force_p_smooth[:, 0:n_plot_smpls]), 
                     force_real=force_blue_int,
                     timestep=ws_hys_tst.ds.timestep)
```
