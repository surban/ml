```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt

from ml.simple.vargp import VarGP
```

```python
def myfunc(x):
    mean = np.sin(x)
    std = 0.1 + 0.2 * np.abs(np.sin(x))
    std = np.where(np.abs(np.sin(x)) > 0.9, 0.15, 0.03)
    val = mean + np.random.normal(size=x.shape) * std
    return val
```

```python
n_samples = 300
max_x = 10
x = np.random.uniform(low=0, high=max_x, size=n_samples)
x = x[np.logical_or(x<6, x>9)]
y = myfunc(x)

plt.plot(x,y, '.')
plt.xlim(0, max_x)
plt.ylim(-1.5, 1.5)
```

```python
vgp = VarGP(x[np.newaxis, :], y, optimize_restarts=5, gp_parameters={'rbf_variance': None}, var_normalize_y=False)
print vgp
```

```python
figsize(20,10)
tst_x = np.linspace(0, max_x, 100)
vgp.plot(x[np.newaxis, :], y, tst_x[np.newaxis, :], ylim=(-2.0, 2.0))
```
