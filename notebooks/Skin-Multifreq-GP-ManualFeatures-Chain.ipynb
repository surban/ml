```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
import GPy as gpy

import ml.apps.skin.timeseries
from ml.apps.skin.timeseries import *
from ml.apps.skin.workingset import *
from ml.simple.conditional_chain import *
from ml.datasets.skin import SkinDataset

ml.apps.skin.timeseries.max_force = 10
```

## Dataset

```python
force_min = -0.2
force_max = 10.0

ws_linear_bidir = SkinWorkingset("mf_linear_bidir_10", taxel=(0,6), force_min=force_min, force_max=force_max)
ws_linear_raising = SkinWorkingset("mf_linear_raising_10", taxel=(0,6), force_min=force_min, force_max=force_max)

ws_random_raising = SkinWorkingset("mf_random_raising_10", taxel=(0,6), force_min=force_min, force_max=force_max)
ws_random_bidir = SkinWorkingset("mf_random_bidir_10", taxel=(0,6), force_min=force_min, force_max=force_max)

ws_hysteresis_10 = SkinWorkingset("mf_hysteresis_10", taxel=(0,6), force_min=force_min, force_max=force_max)
ws_hysteresis_50 = SkinWorkingset("mf_hysteresis_50", taxel=(0,6), force_min=force_min, force_max=force_max)
ws_hysteresis_100 = SkinWorkingset("mf_hysteresis_100", taxel=(0,6), force_min=force_min, force_max=force_max)
```

## Training of hysteresis-free regressor

```python
ws_trn = ws_linear_raising

def select_features(ws, skin):    
    return ws.select_frequencies([794328, 2511886], skin)

figsize(8,4)
plt.plot(ws_trn.force['trn'][:, 0], 'k', label="force")
ftrs = select_features(ws_trn, ws_trn.skin['trn'])[:, :, 0]
for f in range(ftrs.shape[0]):
    plt.plot(10 * ftrs[f, :], label=str(f))
plt.legend()
```

```python
X = select_features(ws_trn, ws_trn.flat_skin['trn']).T
Y = ws_trn.flat_force['trn'][:, np.newaxis]

print X.shape

n_features = X.shape[1]
kernel = gpy.kern.rbf(n_features, ARD=False)
#kernel = gpy.kern.linear(ws_trn.n_features, ARD=True)
gp = gpy.models.GPRegression(X, Y, kernel)

gp.optimize_restarts()
#gp.randomize()
#gp.optimize()
print gp
```

## Evaluation

```python
def evaluate_on_ws(gp, ws, prt, phase=False):
    n_plot_samples = 30

    force = ws.force[prt]
    skin = select_features(ws, ws.skin[prt])
    valid = ws.valid[prt]

    def gp_predict(skin):
        force, force_conf, _, _ = gp.predict(skin.T, full_cov=False)
        return force[:, 0], force_conf[:, 0]

    p_force, p_force_conf = ws.predict_multicurve(gp_predict, force, skin, valid, what='force')
    print "MSE: ", multistep_error(p_force, force, valid, mean_err=True)
    
    ml.apps.skin.timeseries.max_force = 10
    if phase:
        plot_multicurve_error(force[:, 0:n_plot_samples], p_force[:, 0:n_plot_samples], valid[:, 0:n_plot_samples])
    else:
        plot_multicurve_time(force[:, 0:n_plot_samples], None, valid[:, 0:n_plot_samples], 
                             force_predicted=p_force[:, 0:n_plot_samples], force_predicted_conf=p_force_conf[:, 0:n_plot_samples])    
```

```python
figsize(16, 2)
evaluate_on_ws(gp, ws_linear_bidir, 'trn')
```

```python
figsize(16, 8)
evaluate_on_ws(gp, ws_hysteresis_10, 'trn')
```

```python
figsize(16, 16)
evaluate_on_ws(gp, ws_hysteresis_10, 'trn', phase=True)
```

## Infer internal force from true force

```python
ws_hys_tst = ws_hysteresis_100
prt = 'tst'

force, skin, valid = ws_hys_tst.force[prt], ws_hys_tst.skin[prt], ws_hys_tst.valid[prt]
internal_force, _ = ws_hys_tst.predict_multicurve(gp_predict, force, select_features(ws_hys_tst, skin), valid, what='force')

discrete_force = ws_hys_trn.to_discrete_force(force)
discrete_internal_force = ws_hys_trn.to_discrete_force(internal_force)

discrete_internal_force_p = ws_hys_tst.predict_multicurve(lambda f: sc.most_probable_states(f)[0], 
                                                          discrete_force, discrete_internal_force, valid)
print "MSE: ", multistep_error(ws_hys_trn.from_discrete_force(discrete_internal_force_p), 
                               ws_hys_trn.from_discrete_force(discrete_internal_force), valid, mean_err=True)

n_plot_smpls = 42
figsize(18,14)
ml.apps.skin.timeseries.max_force = 10
force_real = ws_hys_trn.from_discrete_force(discrete_force[:, 0:n_plot_smpls])
plot_multicurve_time(ws_hys_trn.from_discrete_force(discrete_internal_force[:, 0:n_plot_smpls]), None, valid[:, 0:n_plot_smpls], 
                     force_predicted=ws_hys_trn.from_discrete_force(discrete_internal_force_p[:, 0:n_plot_smpls]), 
                     force_real=force_real,
                     timestep=ws_hys_tst.ds.timestep)
```

## Infer true force from internal force (without smoothing prior)

```python
ws_hys_tst = ws_hysteresis_100
prt = 'val'

force, skin, valid = ws_hys_tst.force[prt], ws_hys_tst.skin[prt], ws_hys_tst.valid[prt]
internal_force, _ = ws_hys_tst.predict_multicurve(gp_predict, force, select_features(ws_hys_tst, skin), valid, what='force')

discrete_force = ws_hys_trn.to_discrete_force(force)
discrete_internal_force = ws_hys_trn.to_discrete_force(internal_force)

discrete_force_p, discrete_force_log_prob = ws_hys_tst.predict_multicurve(lambda f: sc.infer_inputs(f), 
                                                                          discrete_force, discrete_internal_force, valid, what='force')
discrete_force_prob = np.exp(discrete_force_log_prob)
print "MSE inferred: ", multistep_error(ws_hys_trn.from_discrete_force(discrete_force_p), 
                                        ws_hys_trn.from_discrete_force(discrete_force), valid, mean_err=True)
print "MSE internal: ", multistep_error(ws_hys_trn.from_discrete_force(discrete_internal_force), 
                                        ws_hys_trn.from_discrete_force(discrete_force), valid, mean_err=True)

n_plot_smpls = 42
figsize(18,14)
ml.apps.skin.timeseries.max_force = 10
force_blue = ws_hys_trn.from_discrete_force(discrete_internal_force[:, 0:n_plot_smpls])
plot_multicurve_time(ws_hys_trn.from_discrete_force(discrete_force[:, 0:n_plot_smpls]), None, valid[:, 0:n_plot_smpls], 
                     force_predicted=ws_hys_trn.from_discrete_force(discrete_force_p[:, 0:n_plot_smpls]), 
                     force_prob=discrete_force_prob, 
                     force_prob_extents=ws_hys_trn.from_discrete_force(np.asarray([0, discrete_force_prob.shape[0]])),
                     force_prob_max=0.3,
                     force_real=force_blue,
                     timestep=ws_hys_tst.ds.timestep)
```

## Infer true force from internal force (with smoothing prior)

```python
sigma = 10

force, skin, valid = ws_hys_tst.force[prt], ws_hys_tst.skin[prt], ws_hys_tst.valid[prt]
internal_force, _ = ws_hys_tst.predict_multicurve(gp_predict, force, select_features(ws_hys_tst, skin), valid, what='force')

discrete_force = ws_hys_trn.to_discrete_force(force)
discrete_internal_force = ws_hys_trn.to_discrete_force(internal_force)

discrete_force_p_smooth = ws_hys_tst.predict_multicurve(lambda f: sc.most_probable_inputs(f, sigma=sigma)[0], 
                                                        discrete_force, discrete_internal_force, valid, what='force')
print "MSE inferred smooth: ", multistep_error(ws_hys_trn.from_discrete_force(discrete_force_p_smooth), 
                                               ws_hys_trn.from_discrete_force(discrete_force), valid, mean_err=True)
print "MSE inferred raw:    ", multistep_error(ws_hys_trn.from_discrete_force(discrete_force_p), 
                                               ws_hys_trn.from_discrete_force(discrete_force), valid, mean_err=True)
print "MSE internal:        ", multistep_error(ws_hys_trn.from_discrete_force(discrete_internal_force), 
                                               ws_hys_trn.from_discrete_force(discrete_force), valid, mean_err=True)

n_plot_smpls = 42
figsize(18,14)
ml.apps.skin.timeseries.max_force = 10
force_blue = ws_hys_trn.from_discrete_force(discrete_force_p[:, 0:n_plot_smpls])
plot_multicurve_time(ws_hys_trn.from_discrete_force(discrete_force[:, 0:n_plot_smpls]), None, valid[:, 0:n_plot_smpls], 
                     force_predicted=ws_hys_trn.from_discrete_force(discrete_force_p_smooth[:, 0:n_plot_smpls]), 
                     force_real=None,
                     timestep=ws_hys_tst.ds.timestep)
```
