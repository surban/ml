### Learn next step prediction data

```python
def build_nextstep_ds(vel, phase, freq, trq=None):
    if trq is not None:
        X = np.vstack((phase[1:], freq[1:], trq[1:], vel[0:-1]))
    else:
        X = np.vstack((phase[1:], freq[1:], vel[0:-1]))
    Y = vel[1:]
    return X, Y
```

### predict multiple steps

```python
def ensure_multismpl(x):
    if x.ndim == 2:
        return x
    else:
        return x[:, np.newaxis]

def multistep_predict(gp, vel0, phase, freq, resync_every=0, vel_ref=None):
    phase = ensure_multismpl(phase)
    freq = ensure_multismpl(freq)
    if vel_ref is not None:
        vel_ref = ensure_multismpl(vel_ref)
    
    steps = np.shape(phase)[0]
    smpls = np.shape(phase)[1]
    
    vel = np.zeros((steps, smpls))
    vel_var = np.zeros((steps, smpls))
    vel[0, :] = vel0
    for t in range(1, steps):
        X = np.vstack((phase[t, :], freq[t, :], vel[t-1, :]))       
        P_Y, P_Y_var = gp.predict(X.T)
        vel[t, :] = P_Y.T[0, :]
        vel_var[t, :] = P_Y_var[0, :]
        
        if resync_every != 0 and t % resync_every == 0:
            vel[t, :] = vel_ref[t, :]            
        
    return vel, vel_var
```

```python
n=0
resync = 0
P_vel, P_vel_var = multistep_predict(gp, vel[n][0], phase[n], freq[n], resync_every=resync, vel_ref=vel[n])

plt.clf()
plt.hold(True)
plt.plot(P_vel, 'r')
plt.plot(P_vel + 3*np.sqrt(P_vel_var), color='gray')
plt.plot(P_vel - 3*np.sqrt(P_vel_var), color='gray')
#for t in range(0, 600, resync):
#    plt.axvline(t)
plt.plot(vel[n], 'b', linewidth=0.1)
```

```python
def multistep_predict_uncertain(gp, vel0, phase, freq, resync_every=0, vel_ref=None):
    phase = ensure_multismpl(phase)
    freq = ensure_multismpl(freq)
    if vel_ref is not None:
        vel_ref = ensure_multismpl(vel_ref)
    
    steps = np.shape(phase)[0]
    smpls = np.shape(phase)[1]
    
    vel = np.zeros((steps, smpls))
    vel_var = np.zeros((steps, smpls))
    vel[0, :] = vel0
    vel_var[0, :] = 0
    for t in range(1, steps):
        X_mu = np.zeros((smpls, 3))
        X_mu[:, 0] = phase[t, :]
        X_mu[:, 1] = freq[t, :]
        X_mu[:, 2] = vel[t-1, :]
        X_covar = np.zeros((smpls, 3, 3))
        X_covar[:, 2, 2] = vel_var[t-1, :]
        
        P_Y, P_Y_var = gp.predict_uncertain(X_mu, X_covar)
        vel[t, :] = P_Y[:, 0]
        vel_var[t, :] = P_Y_var[:, 0, 0]
        
        if resync_every != 0 and t % resync_every == 0:
            vel[t, :] = vel_ref[t, :]  
            vel_var[t, :] = 0
        
    return vel, vel_var
```

```python
n=0
resync = 0
P_vel, P_vel_var = multistep_predict_uncertain(gp, vel[n][0], phase[n], freq[n], resync_every=50, vel_ref=vel[n])

plt.clf()
plt.hold(True)
plt.plot(P_vel, 'r')
plt.plot(P_vel + np.sqrt(P_vel_var), color='gray')
plt.plot(P_vel - np.sqrt(P_vel_var), color='gray')
#for t in range(0, 600, resync):
#    plt.axvline(t)
plt.plot(vel[n], 'b', linewidth=0.1)
plt.ylim(-3, 3)
```
