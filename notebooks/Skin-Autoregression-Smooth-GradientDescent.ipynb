```python
force_min = 0
force_step = 0.1
force_max = 25
skin_min = 0
skin_step = 0.02
skin_max = 2

ds = SkinDataset("raising_small")
ds.print_statistics()
```

```python
taxel = (1,1)
n_curves = None

X, Z = build_nextstep_data(ds, 'train', taxel, n_curves=n_curves)
print "Trainingset size: %d steps" % X.shape[1]

Xval, Zval = build_nextstep_data(ds, 'validation', taxel)
print "Validationset size: %d steps" % Xval.shape[1]

Xtst, Ztst = build_nextstep_data(ds, 'test', taxel)
print "Testset size: %d steps" % Xtst.shape[1]

test_curves = ds.record('test', (1,1), range(20))
test_force, test_skin, test_valid = build_multicurve(test_curves)
```

# Vanilla autoregression

```python
tr = SmoothTableRegression([force_min, skin_min], 
                           [force_step, skin_step], 
                           [force_max, skin_max])
print "Number of weights: %d" % tr.weight_matrix.size
```

```python
lsmr_err = 0.00541646634298
```

```python
tr.train_lsmr(X, Z)
```

```python
Zval_p = tr.predict(Xval)
lsmr_err = np.sqrt(np.mean((Zval - Zval_p)**2))
print "One step prediction sqrt(MSE): ", lsmr_err
```

### Training using gradient descent

```python
batch_size = 10000
target_err = lsmr_err*2

tr_gd = SmoothTableRegression([force_min, skin_min], 
                              [force_step, skin_step], 
                              [force_max, skin_max])
#tr_gd._weights = np.random.random(size=tr._weights.shape)
print "Number of weights: %d" % tr_gd.weight_matrix.size

n_batches = int(X.shape[1] / batch_size)
batch_no = 0

print "Number of batches: %d" % n_batches

def weight_wrapper(w):
    x = X[:, batch_no*batch_size : (batch_no+1)*batch_size]
    z = Z[batch_no*batch_size : (batch_no+1)*batch_size]
    dEdF = tr_gd.predict(x) - z   
    x_grad, w_grad = tr_gd.gradient(x)
    dFdW = w_grad.toarray()
    dEdW = np.dot(dFdW, dEdF)
    #print "Gradient:", dEdW
    return dEdW
```

```python
gd = GradientDescent(tr_gd._weights, weight_wrapper, steprate=0.01, momentum=0.0)
for i, info in enumerate(gd):
    batch_no = (batch_no + 1) % n_batches
    
    if i % 100 == 0:
        Zval_p = tr_gd.predict(Xval)
        err = np.sqrt(np.mean((Zval - Zval_p)**2))
        print "%d: one step prediction sqrt(MSE): %g" % (i, err)       
        if err < target_err:
            break
```
