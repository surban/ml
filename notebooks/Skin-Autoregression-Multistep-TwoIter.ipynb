## Batch approach

```python
new_batch_iters = 10

his_failed = ParameterHistory(min_iters=1000)
his_ns = ParameterHistory(show_progress=False)
his_ms = ParameterHistory(show_progress=False)            

msws = SkinMultistepWorkingset(ws, tr, ws.force['trn'], ws.skin['trn'], ws.valid['trn'], 50)
nsws = SkinNextstepWorkingset(ws, tr, None, None)
opt = RmsProp(tr.weights, nsws.gradient_wrapper, steprate=1e-6)

msws.batch = 0
f, v, s = msws.get_fvs()
s_p = multistep_predict(tr.predict, f, v, s[0, :])
nsws.ns_in, nsws.ns_skin = build_nextstep_data_from_multicurve(f, s_p, v, s)

for iter, info in enumerate(opt):
    if iter % new_batch_iters == 0:
        msws.next_batch()
        f, v, s = msws.get_fvs()
        s_p = multistep_predict(tr.predict, f, v, s[0, :])
        nsws.ns_in, nsws.ns_skin = build_nextstep_data_from_multicurve(f, s_p, v, s)

    if iter % 100 == 0:        
        err = ws.error(tr)
        his_ns.add(iter, tr.weights, err['ns_trn'], err['ns_val'], err['ns_tst'])
        his_ms.add(iter, tr.weights, err['ms_trn'], err['ms_val'], err['ms_tst'])
        his_failed.add(iter, tr.weights, len(err['failed_trn']), len(err['failed_val']), len(err['failed_tst']))                  
        
        plt.clf()
        his_failed.plot(final=False, logscale=False)
        plt.show()           
    
```

## Full multistep Debugging

```python
def plot_grad(tr, grad, rng=None):
    grad = np.reshape(grad, tuple(tr._elems), 'F')
    
    if not rng:
        rng = np.amax(np.fabs(grad))

    img = plt.imshow(grad.T, origin='lower', interpolation='nearest', cmap='PuOr',
                     extent=(tr._mins[0], tr._maxs[0], tr._mins[1], tr._maxs[1]),
                     aspect='auto')
    img.set_clim(-rng, rng)
    plt.colorbar()
    plt.xlabel("dim 0")
    plt.ylabel("dim 1")
```

```python
tr.weights = start_weights.copy()
err = ws.error(tr)
failed_force = ws.force['trn'][:, err['failed_trn']]
failed_skin = ws.skin['trn'][:, err['failed_trn']]
failed_valid = ws.valid['trn'][:, err['failed_trn']]
print failed_force.shape
```
