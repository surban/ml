```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
from scipy.stats import norm
```

```python
norm.cdf(0.4, loc=0, scale=1)
```

```python
x=np.arange(-1, 102, 1)
plot(x, norm.cdf(x, loc=50, scale=10))
```

```python
cdf = lambda x: norm.cdf(x, loc=50, scale=10)

low_lim = 20
high_lim = 80

pos = np.arange(low_lim, high_lim+1)
cdt = cdf(pos)
pdt = np.diff(cdt)
pdt[0] += cdf(low_lim)
pdt[-1] += 1-cdf(high_lim)
pos = pos[0:-1]

plot(pos, pdt)
print np.sum(pdt)
```
