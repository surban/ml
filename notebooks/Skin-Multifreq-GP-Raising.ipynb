```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
import GPy as gpy

import ml.apps.skin.timeseries
from ml.apps.skin.timeseries import *
from ml.apps.skin.workingset import *
from ml.simple.conditional_chain import *
from ml.datasets.skin import SkinDataset
from ml.common.util import ParameterHistory
from ml.common.test import check_gradient, check_directional_gradient

ml.apps.skin.timeseries.max_force = 10
```

## Dataset

```python
ws = SkinWorkingset("mf_raising_small1", taxel=(0,6))
ws_bidir = SkinWorkingset("mf_small1", taxel=(0,6))
```

## Training

```python
kernel = gpy.kern.rbf(ws.n_features, ARD=False)
X = ws.flat_skin['trn'].T
Y = ws.flat_force['trn'][:, np.newaxis]
gp = gpy.models.GPRegression(X, Y, kernel)

gp.optimize()
print gp
```

## Evaluation on raising dataset

```python
prt = 'tst'
n_plot_samples = 30

force = ws.force[prt]
skin = ws.skin[prt]
valid = ws.valid[prt]

def gp_predict(skin):
    force, force_conf, _, _ = gp.predict(skin.T, full_cov=False)
    return force[:, 0], force_conf[:, 0]

p_force, p_force_conf = ws.predict_multicurve(gp_predict, force, skin, valid, what='force')
print "MSE: ", multistep_error(p_force, force, valid, mean_err=True)

ml.apps.skin.timeseries.max_force = 12
figsize(20,10)
plot_multicurve_time(force[:, 0:n_plot_samples], None, valid[:, 0:n_plot_samples], 
                     force_predicted=p_force[:, 0:n_plot_samples], force_predicted_conf=p_force_conf[:, 0:n_plot_samples])
```

## Evaluation on bidirectional dataset

```python
prt = 'tst'
n_plot_samples = 30

force = ws_bidir.force[prt]
skin = ws_bidir.skin[prt]
valid = ws_bidir.valid[prt]

def gp_predict(skin):
    force, force_conf, _, _ = gp.predict(skin.T, full_cov=False)
    return force[:, 0], force_conf[:, 0]

p_force, p_force_conf = ws_bidir.predict_multicurve(gp_predict, force, skin, valid, what='force')
print "MSE: ", multistep_error(p_force, force, valid, mean_err=True)

ml.apps.skin.timeseries.max_force = 10
figsize(20,10)
plot_multicurve_time(force[:, 0:n_plot_samples], None, valid[:, 0:n_plot_samples], 
                     force_predicted=p_force[:, 0:n_plot_samples], force_predicted_conf=p_force_conf[:, 0:n_plot_samples])
```
