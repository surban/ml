```python
def my_predict(X, w, sv, sigma):
    sum = np.zeros(X.shape[0])
    for i in range(len(w)):
        d = X-sv[i,:]
        d2 = d * d
        d2sum = np.sum(d2, axis=1)
        K = np.exp(-d2sum / (2*sigma**2))
        sum += w[i]*K
    return sum

w = np.random.random((20000,))
sv = np.random.random((20000,784))
sigma = 1

s=my_predict(gp.as_numpy_array(tcfg.X[0:100]), w, sv, sigma)
```
