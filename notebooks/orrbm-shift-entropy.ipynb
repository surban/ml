```python
%load_ext autoreload
%autoreload 2

# workaround for Theano bug
import os
# turn off for training reference_deep_svm
os.environ['BREZE_PARAMETERSET_DEVICE'] = 'cpu'

from rbm.rbm import train_rbm, RestrictedBoltzmannMachine
from rbm.config import TrainingConfiguration
from rbm.ais import AnnealedImportanceSampler
from common.util import flatten_samples, unflatten_samples_like
import apps.mnist_reference_dropout as ref
import apps.generate_letters
import rbm.util
import rbm.sampling
import rbm.accuracy
import rbm.orrbm
import common
import common.util
import common.progress
import gnumpy as gp
import numpy as np
import pylab
import os


gp.expensive_check_probability = 0
```

```python
n_vis=784
n_hid=5000
init_weight_sigma = 4 * np.sqrt(6. / (n_hid + n_vis))
tcfg = TrainingConfiguration(dataset='mnistv',
                             n_vis=n_vis, n_hid=n_hid,
                             batch_size=20,
                             n_gibbs_steps=15,
                             epochs=30,
                             step_rate=0.1,
                             use_pcd=True,
                             binarize_data='round',
                             initial_momentum=0, final_momentum=0, 
                             use_final_momentum_from_epoch=0,
                             weight_cost=0,
                             init_method='uniform', init_weight_sigma=init_weight_sigma, init_bias_sigma=0,
                             seed=1)
rbm.util.enter_rbm_plot_directory(tcfg)

epoch = tcfg.epochs - 1
myrbm = RestrictedBoltzmannMachine(20, n_vis, n_hid, 0)
rbm.util.load_parameters(myrbm, epoch)

X2 = tcfg.X.reshape((tcfg.X.shape[0], 28, 28))

ref_predict = ref.build_predictor()

ref_Z=common.util.map(gp.as_numpy_array(tcfg.X), 1000, ref_predict)
ref_TZ=common.util.map(gp.as_numpy_array(tcfg.TX), 1000, ref_predict)
ref_acc = 1 - common.util.classification_error(ref_Z, tcfg.Z)
ref_tacc = 1 - common.util.classification_error(ref_TZ, tcfg.TZ)
print "Reference classification error on whole training set:  %g" % (1-ref_acc)
print "Reference classification error on whole test set:      %g" % (1-ref_tacc)
```

```python
def or_sample_with_shift(i, j, x_shift, y_shift):
    return rbm.orrbm.or_sample_with_shift(X2[i:i+1,:,:], X2[j:j+1,:,:], x_shift, y_shift)
```

## Calculate cross entropy of different shifts

```python
points = 10
print rbm.orrbm.cross_entropy(myrbm, ors, points, gen_x_shift, gen_y_shift, iters=20, k=10, beta=2)
print rbm.orrbm.cross_entropy(myrbm, ors, points, gen_x_shift-4, gen_y_shift, iters=20, k=10, beta=2)
```

```python
points = 10
Hs = []
max_xs = 28
for xs in range(max_xs):
    common.progress.status(xs, max_xs, "Calculating cross entropy")    
    H = rbm.orrbm.cross_entropy(myrbm, ors, points, xs, gen_y_shift, iters=10, k=10, beta=2)
    H = gp.as_numpy_array(H)
    Hs.append(H[0])
Hs = np.asarray(Hs)
plt.plot(Hs)
```

```python
np.argmax(Hs)
```

```python
infer_x_shift = np.argmax(Hs)

x, y = rbm.orrbm.or_infer_with_shift(myrbm, ors, infer_x_shift, gen_y_shift, iters=20, k=10, beta=2)
common.util.ipy_plot_samples([x[0], y[0]], twod=True)

zx=ref_predict(gp.as_numpy_array(flatten_samples(x)))
zy=ref_predict(gp.as_numpy_array(flatten_samples(y)))
print "x: %d   y: %d" % (zx, zy)
```

## Heatmap of shifts

```python
points = 1
max_xs = rbm.orrbm.width
max_ys = rbm.orrbm.base_y * 2
x_shifts, y_shifts = np.meshgrid(range(max_xs), range(max_ys))
Hs = np.zeros_like(x_shifts)
for i in range(x_shifts.shape[0]):
    common.progress.status(i, x_shifts.shape[0], "Calculating cross entropy") 
    for j in range(x_shifts.shape[1]):
        x_shift = x_shifts[i,j]
        y_shift = y_shifts[i,j]
        H = rbm.orrbm.cross_entropy(myrbm, ors, points, x_shift, y_shift, iters=10, k=10, beta=2)
        H = gp.as_numpy_array(H)[0]
        Hs[i,j] = H
        
        
```
