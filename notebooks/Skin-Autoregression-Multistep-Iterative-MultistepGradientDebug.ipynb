## Finetuning using iterative full prediction and gradient descent

```python
restart_steps = 30
next_step_iters = 10

#tr.weights = start_weights.copy()
#tr.weights = np.zeros(tr.weights.shape)
figsize(12,6)

his_failed = ParameterHistory(min_iters=1000)
his_ns = ParameterHistory(show_progress=False)
his_ms = ParameterHistory(show_progress=False)            

msaws = SkinMultistepAlternativeWorkingset(ws, tr, ws.force['trn'], ws.skin['trn'], ws.valid['trn'], restart_steps,
                                           fallback_gradient_norm=5, random_sample_shift=True)
opt = GradientDescent(tr.weights, msaws.gradient_wrapper, steprate=1e-3)

msaws.start_step = 0
for iter, info in enumerate(opt):
    #print "max weight:", np.max(tr.weights)
    #print "max gradient:", np.max(msaws.gradient_wrapper(None))
    #assert not np.any(np.isnan(tr.weights))
    
    if iter % next_step_iters == 0:
        msaws.next_batch()
        #print "Start step: ", msaws.start_step

    if iter % 100 == 0:        
        err = ws.error(tr)
        his_ns.add(iter, tr.weights, err['ns_trn'], err['ns_val'], err['ns_tst'])
        his_ms.add(iter, tr.weights, err['ms_trn'], err['ms_val'], err['ms_tst'])
        his_failed.add(iter, tr.weights, len(err['failed_trn']), len(err['failed_val']), len(err['failed_tst']))                  
        
        plt.clf()
        plt.subplot(1,3,1); his_ns.plot(final=False, logscale=True)
        plt.subplot(1,3,2); his_ms.plot(final=False, logscale=True)
        plt.subplot(1,3,3); his_failed.plot(final=False, logscale=False)       
        plt.show()        
        
        plt.figure()
        my_s_p = restarting_multistep_predict(tr.predict, my_f, my_v, my_s, restart_steps)
        plot_multicurve_time(my_f, my_s, my_v, my_s_p, timestep=ws.ds.timestep)        
        plt.show()                    
        
        plt.figure()
        #figsize(12,6)
        tr.plot2d_relative(rel_dim=1, rng=0.5)
        plt.xlabel("force")
        plt.ylabel("skin")
        plt.show()
```

```python
tr.weights = end_weights.copy()
```

```python
tr.weights = end_weights.copy()
figsize(12,6)
next_step_iters = 10

his_failed = ParameterHistory(min_iters=1000)
his_ns = ParameterHistory(show_progress=False)
his_ms = ParameterHistory(show_progress=False)            

msaws = SkinMultistepWorkingset(ws, tr, ws.force['trn'], ws.skin['trn'], ws.valid['trn'], 10)
opt = GradientDescent(tr.weights, msaws.gradient_wrapper, steprate=1e-6)

msaws.start_step = 0
for iter, info in enumerate(opt):
    if iter % next_step_iters == 0:
        msaws.next_batch()

    if iter % 1 == 0:        
        err = ws.error(tr)
        his_ns.add(iter, tr.weights, err['ns_trn'], err['ns_val'], err['ns_tst'])
        his_ms.add(iter, tr.weights, err['ms_trn'], err['ms_val'], err['ms_tst'])
        his_failed.add(iter, tr.weights, len(err['failed_trn']), len(err['failed_val']), len(err['failed_tst']))                  
        
        plt.clf()
        plt.subplot(1,3,1); his_ns.plot(final=False, logscale=False)
        plt.subplot(1,3,2); his_ms.plot(final=False, logscale=False)
        plt.subplot(1,3,3); his_failed.plot(final=False, logscale=False)       
        plt.show()                 
               
np.savez_compressed("multistep-restart-finetuned.npz", weights=tr.weights)     
```
