```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import scipy.io
import GPy 
```

## Onedimensional input

```python
X = np.array([[1],
              [2],
              [3]])
Y = np.array([[1.0, 4.0],
              [2.5, 5.0],
              [3.1, 6.5]])

kernel = GPy.kern.RBF(input_dim=X.shape[1], ARD=False)
gp = GPy.models.GPRegression(X, Y, kernel)
gp.optimize()
gp.plot()
display(gp)
```

```python
Xnew = np.array([[1.5], [2.5]])

Ynew_mu, Ynew_var = gp.predict(Xnew, full_cov=False)
print "mu=\n", Ynew_mu
print "var=\n", Ynew_var
```

```python
Xnew_mu = np.array([[1.5], [2.5]])
Xnew_cov = np.array([[[0.01]], [[0.0000001]]])

Ynew_mu, Ynew_covar = gp.predict_uncertain(Xnew_mu, Xnew_cov)
print "mu=\n", Ynew_mu
print "covar=\n", Ynew_covar
```

## Multidimensional input

```python
X = np.array([[1, 1.1],
              [2, 2.3],
              [3, 3.3]])
Y = np.array([[1.0, 4.0],
              [2.5, 5.0],
              [3.1, 6.5]])

kernel = GPy.kern.RBF(input_dim=X.shape[1], ARD=False)
gp = GPy.models.GPRegression(X, Y, kernel)
gp.optimize()
gp.plot()
display(gp)
```

```python
Xnew = np.array([[1.5, 1.6], 
                 [2.5, 2.6]])

Ynew_mu, Ynew_var = gp.predict(Xnew, full_cov=False)
print "mu=\n", Ynew_mu
print "var=\n", Ynew_var
```

```python
Xnew_mu = np.array([[1.5, 1.6], 
                    [2.5, 2.6]])
Xnew_cov = np.array([[[0.1, 0.0],
                      [0.0, 0.1]], 
                     [[1e-6, 0.00],
                      [0.00, 1e-6]]])

Ynew_mu, Ynew_covar = gp.predict_uncertain(Xnew_mu, Xnew_cov)
print "mu=\n", Ynew_mu
print "covar=\n", Ynew_covar
```
