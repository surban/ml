## Dataset

```python
#ws = SkinWorkingset("bidir_small", (1, 1), discrete_force_states=100, discrete_skin_states=100)
ws = SkinWorkingset("bidir_small", (7, 3), discrete_force_states=400, discrete_skin_states=100)

sc = ConditionalChain(ws.discrete_skin_states, ws.discrete_force_states)
```

## Transitions

```python
figsize(16,4)
interact(lambda pstate: sc.plot_transitions(ws.discrete_skin['trn'], ws.discrete_force['trn'], ws.valid['trn'], pstate),
         pstate=widgets.IntSliderWidget(min=0, max=sc.n_system_states, step=1, value=20));
```

## GP visualization

```python
figsize(16, 8)
interact(lambda pstate: sc.apply_gp_to_state(ws.discrete_skin['trn'], ws.discrete_force['trn'], ws.valid['trn'], pstate=pstate,
                                             do_plots=True,
                                             mngful_dist=3, cutoff_stds=1.0,
                                             gp_kernel=lambda: gpy.kern.rbf(1), gp_normalize_x=False, gp_normalize_y=True, 
                                             gp_optimize=True, 
                                             gp_parameters={'noise_variance': None, 'rbf_variance': None}),                         
         pstate=widgets.IntSliderWidget(min=0, max=sc.n_system_states, step=1, value=43));
```

## Training

```python
sc.train_gp_var(ws.discrete_skin['trn'], ws.discrete_force['trn'], ws.valid['trn'], 
                cutoff_stds=1.0, mngful_dist=int(0.6 / ws.force_step),
                gp_kernel=lambda: gpy.kern.rbf(1), gp_normalize_x=False, gp_normalize_y=True, 
                gp_optimize=True, gp_parameters={})
```

## Evaluation

### on test set

```python
prt = 'tst'
force, skin, valid = ws.discrete_force[prt], ws.discrete_skin[prt], ws.valid[prt]
skin_p = ws.predict_multicurve(lambda f: sc.most_probable_states(f)[0], force, skin, valid)
print "MSE: ", multistep_error(ws.from_discrete_skin(skin_p), ws.from_discrete_skin(skin), valid, mean_err=True)

n_plot_smpls = 42
figsize(18,20)
plot_multicurve_time(ws.from_discrete_force(force[:, 0:n_plot_smpls]), ws.from_discrete_skin(skin[:, 0:n_plot_smpls]), valid[:, 0:n_plot_smpls], 
                     ws.from_discrete_skin(skin_p[:, 0:n_plot_smpls]), timestep=ws.ds.timestep)
```

### on subset of training set

```python
prt = 'trn'
start_smpl = 400
n_plot_smpls = 42

force, skin, valid = (ws.discrete_force[prt][:, start_smpl:start_smpl+n_plot_smpls], 
                      ws.discrete_skin[prt][:, start_smpl:start_smpl+n_plot_smpls], 
                      ws.valid[prt][:, start_smpl:start_smpl+n_plot_smpls])
skin_p = ws.predict_multicurve(lambda f: sc.most_probable_states(f)[0], force, skin, valid)
print "MSE: ", multistep_error(ws.from_discrete_skin(skin_p), ws.from_discrete_skin(skin), valid, mean_err=True)

figsize(18,20)
plot_multicurve_time(ws.from_discrete_force(force), ws.from_discrete_skin(skin), valid, 
                     ws.from_discrete_skin(skin_p), timestep=ws.ds.timestep)
```
