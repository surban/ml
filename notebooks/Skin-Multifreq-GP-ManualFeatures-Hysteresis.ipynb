## Dataset

```python
ws_linear_bidir = SkinWorkingset("mf_linear_bidir_10", taxel=(0,6))
ws_linear_raising = SkinWorkingset("mf_linear_raising_10", taxel=(0,6))

ws_random_raising = SkinWorkingset("mf_random_raising_10", taxel=(0,6))
ws_random_bidir = SkinWorkingset("mf_random_bidir_10", taxel=(0,6))
ws_hysteresis = SkinWorkingset("mf_hysteresis_10", taxel=(0,6))
```

## Training

```python
def select_features(skin):    
    return skin[[40, -2], ...]

ws_trn = ws_linear_raising

plt.plot(ws_trn.force['trn'][:, 0], 'k', label="force")
ftrs = select_features(ws_trn.skin['trn'])[:, :, 0]
for f in range(ftrs.shape[0]):
    plt.plot(10 * ftrs[f, :], label=str(f))
plt.legend()
```

```python
X = select_features(ws_trn.flat_skin['trn']).T
Y = ws_trn.flat_force['trn'][:, np.newaxis]

print X.shape

n_features = X.shape[1]
kernel = gpy.kern.rbf(n_features, ARD=False)
#kernel = gpy.kern.linear(ws_trn.n_features, ARD=True)
gp = gpy.models.GPRegression(X, Y, kernel)

gp.optimize_restarts()
#gp.randomize()
#gp.optimize()
print gp
```

## Evaluation

```python
def evaluate_on_ws(gp, ws, prt, phase=False):
    n_plot_samples = 30

    force = ws.force[prt]
    skin = select_features(ws.skin[prt])
    valid = ws.valid[prt]

    def gp_predict(skin):
        force, force_conf, _, _ = gp.predict(skin.T, full_cov=False)
        return force[:, 0], force_conf[:, 0]

    p_force, p_force_conf = ws.predict_multicurve(gp_predict, force, skin, valid, what='force')
    print "MSE: ", multistep_error(p_force, force, valid, mean_err=True)

    ml.apps.skin.timeseries.max_force = 10
    if phase:
        plot_multicurve_error(force[:, 0:n_plot_samples], p_force[:, 0:n_plot_samples], valid[:, 0:n_plot_samples])
    else:
        plot_multicurve_time(force[:, 0:n_plot_samples], None, valid[:, 0:n_plot_samples], 
                             force_predicted=p_force[:, 0:n_plot_samples], force_predicted_conf=p_force_conf[:, 0:n_plot_samples])    
```

```python
figsize(16, 2)
evaluate_on_ws(gp, ws_linear_raising, 'trn')
```

```python
figsize(16, 8)
evaluate_on_ws(gp, ws_hysteresis, 'trn')
```

```python
figsize(16, 16)
evaluate_on_ws(gp, ws_hysteresis, 'trn', phase=True)
```
